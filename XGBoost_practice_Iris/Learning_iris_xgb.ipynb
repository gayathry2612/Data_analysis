```python
%reset
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn import model_selection
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
dataset = pd.read_csv('iris.csv',header=None)
dataset
```

```python
#y = dataset[:,4]
#x = dataset[:,0:4]
#You cannot hash a dataframe

dataset_values = dataset.values
y = dataset_values[:,4]
x = dataset_values[:,0:4]
```

```python
# encode string class values as integers
label_encoder = LabelEncoder()
label_encoder = label_encoder.fit(y)
label_encoded_y = label_encoder.transform(y)
```

```python
seed = 7 
test_size = 0.33
X_train,X_test,Y_train,Y_test = model_selection.train_test_split(x,label_encoded_y, test_size=test_size,random_state=seed)
```

```python
model = xgb.XGBClassifier()
#fit the model 
model.fit(X_train,Y_train)
print model

"""
Basically I have put in the dataset - train and test into XGB classifier. the output of the XGB classifier gave out 
outputs as below

"""
```

```python
#Time to make predictions
y_predict = model.predict(data=X_test)
predictions = [round(value) for value in y_predict] #To ensure good mapping 
accuracy = accuracy_score(Y_test,predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
