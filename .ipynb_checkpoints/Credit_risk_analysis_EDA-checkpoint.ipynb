```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn as sk
from sklearn import preprocessing as pp
%matplotlib inline
import xgboost as xg
```

```python
data = pd.read_csv('creditcard.csv')
data.info()

#It's a neat dataset - the creators ran a PCA on this dataset
```

```python
#Normalization
y = data["Class"]
time = data["Time"]
data = data.drop('Class',1)
data = data.drop('Time',1)
data["Amount"] = pp.StandardScaler().fit_transform(data['Amount'].values.reshape(-1,1))
```

```python
#Calculating correlation between the variables and amount - Define XGB params
xgb_params = {
    'eta' : 0.05,
    'max_depth':8,
    'subsample':0.7,
    'colsample_bytree':0.7,
    'objective':'binary:logistic',
    'eval_metric': 'auc',
    'silent' : 1
}

dtrain = xg.DMatrix(data,y,feature_names=data.columns.values)
model = xg.train(dict(xgb_params,silent = 0),dtrain,num_boost_round=100)

fig, ax = plt.subplots(figsize=(12,18))
xg.plot_importance(model, height=0.8, ax=ax)
plt.show()
```
